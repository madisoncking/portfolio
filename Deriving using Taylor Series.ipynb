```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Deriving a Taylor series for $\sin(x^{2})$ around $x=0$ using:  

$$
\sin(x) = \sum_{j=0}^{\infty}\frac{(-1)^{j}}{(2j + 1)!}x^{2j +1},
$$



```python
# below is a code which computes the Taylor series of  sin(𝑥^2)  using a for loop.
def tn_approx(x,n):
    tot = 1.
    term = 1.
    for jj in xrange(1,n+1):
        term *= -x**4./((2*jj+1)*(2*jj))
        tot += term
    return x*x*tot
```

```python
# below is a code which computes the Taylor series of sin(x^2) using recursion and a tolerance based while loop approach
def sin_sqd(x,tol):
    tot = 1. # for the total you wish to return 
    aj = 1. # the terms you need to update
    j = 1 # your counter
    xall = -x**4.
    while np.abs(aj)>=tol:
        aj *= xall/((2*j+1)*(2*j))
        tot += aj
        j+=1 
    return x*x*tot 
```

```python
# plotting to view the accuracy
xvals = np.linspace(0, np.pi,int(1e3))
yvalst = np.sin(xvals**2)
yvals1 = [sin_sqd(xval,1e-3) for xval in xvals]
```

```python
plt.plot(xvals,yvalst,color='r', label="$sin(x^2)$")
plt.plot(xvals,yvals1,color='k',ls='--', label="$Taylor \ approx. of \ sin(x^2)$")
plt.legend(loc=(.05,.17))
```
